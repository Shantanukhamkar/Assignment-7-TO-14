## PCA


```python
# importing libraries
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.preprocessing import scale
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
import warnings

warnings.filterwarnings("ignore")
```

```python
# Load dataset
wine = pd.read_csv("/content/wine.csv")
```

```python
wine
```

```python
wine.info()
```

# Clustering without Conducting PCA

Firstly we will conduct clustering without applying PCA

## Hierarchical Clustering

```python
#createing a copy of dataset
```

```python
wine_hc = wine.copy()
```

```python
wine_hc
```

```python
# Standardization
wine_norm_hc = scale(wine.iloc[:,1:])
wine_norm_hc
```

```python
# Create dendrogram
plt.figure(figsize=(20,10))
dendrogram = sch.dendrogram(sch.linkage(wine_norm_hc,method="ward"))
```

Seeing at the dendrogram we can select number of clusters as 3

```python
#create clusters
hc = AgglomerativeClustering(n_clusters=3, affinity = "euclidean", linkage = "complete")
hc.fit(wine_norm_hc)
```

```python
y_hc = hc.fit_predict(wine_norm_hc)
y_hc
```

```python
# adding the clusters to the dataframe
wine_hc['clusterID_hc'] = y_hc
wine_hc.head()
```

```python
#number of datapoints for each clusters
wine_hc['clusterID_hc'].value_counts()
```

```python
#comparing the clusters with the actual categories
pd.crosstab(wine_hc["Type"], wine_hc["clusterID_hc"])
```

## K-Means Clustering

```python
# create a copy 
wine_km = wine.copy()
wine_km
```

```python
#standardization
wine_norm_km = scale(wine_km.iloc[:,1:])
wine_norm_km
```

```python
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i,random_state=0)
  kmeans.fit(wine_norm_km)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title("Elbow Method")
plt.xlabel("No. of Clusters")
plt.ylabel("WCSS")
plt.show()
```

Looking at thr curve we may conclude that the number of clusters shall be taken as 3.

```python
#Build Cluster algorithm
clusters_new = KMeans(n_clusters=3,random_state=42)
clusters_new.fit(wine_norm_km)
```

```python
clusters_new.labels_
```

```python
#assign clusters to the dataset
wine_km["clusterId_km"] = clusters_new.labels_
wine_km.head()
```

```python
#number of datapoints for each clusters
wine_km['clusterId_km'].value_counts()
```

```python
#comparing the clusters with the actual categories
pd.crosstab(wine_km["Type"], wine_km["clusterId_km"])
```

### Conducting PCA

```python
#removing the column of actual categories 
wine1 = wine.iloc[:,1:]
wine1.head()
```

```python
#standardization
wine_normal = scale(wine1)
wine_normal
```

```python
#creating 3 principle components
pca = PCA(n_components = 3)
pca_values = pca.fit_transform(wine_normal)
pca_values
```

```python
#explained variance
var = pca.explained_variance_ratio_
var
```

```python
#cumulative explained variance
var1 = np.cumsum(np.round(var, decimals = 4)*100)
var1
```

```python
pca.components_
```

```python
#variance plot for PCA components obtained scree plot
plt.plot(var1,color="red")
```

```python
#combining the features and the column of actual category
finalwine = pd.concat([pd.DataFrame(pca_values[:,0:3], columns=["pc1","pc2","pc3"]), wine[["Type"]]],axis=1)
finalwine
```

####Clustering after applying PCA

Now we shall conduct clustering using the principal components

### Hierarchical Clustering

```python
#creating a copy
wine1_hc = finalwine.copy()
```

```python
#create dendrogram
plt.figure(figsize=(20,10))
dendrogram1 = sch.dendrogram(sch.linkage(wine1_hc, method="ward"))
```

Seeing at the dendrogram we can select number of clusters as 3.

```python
#create clusters
hc1 = AgglomerativeClustering(n_clusters=3, affinity = "euclidean", linkage = "complete")
hc1.fit(wine1_hc)
```

```python
y_hc1 = hc1.fit_predict(wine1_hc)
y_hc1
```

```python
#adding the clusters to the dataframe
wine1_hc['clusterID_hc'] = y_hc1
wine1_hc.head()
```

```python
#number of datapoints for each clusters
wine1_hc['clusterID_hc'].value_counts()
```

```python
#comparing the clusters with the actual categories
pd.crosstab(wine1_hc["Type"], wine1_hc["clusterID_hc"])
```
